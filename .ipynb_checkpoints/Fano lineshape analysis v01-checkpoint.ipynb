# Function definitions

## Gate lever arm

```python
alpha = 0.07 #eV/V
```

## Line shapes

```python
#df/dE:
def thermal_broadening(E, T):    
    return (1/(4.0*kB*T))*(np.cosh(E/(2.0*kB*T)))**(-2)


#Line shape:
def Breit_Wigner(E, A, Vg, Vg_0, Gamma):
    'in units of 2e2/h'
    return 1/2/A/pi*h*Gamma/((h*Gamma/2)**2+(alpha*e*(Vg-Vg_0)-E)**2)


def Fano(E, A, Vg, Vg_0, Gamma, qRe, qIm):
    'in units of 2e2/h'
    return A*(((alpha*e*(Vg-Vg_0)-E)/(h*Gamma/2)+qRe)**2+qIm**2)/(1+((alpha*e*(Vg-Vg_0)-E)/(h*Gamma/2))**2)


#Integrand:
def CoulIntegrand(E, A, Vg, Vg_0, Gamma, T):
    'the intergrand for the Coulomb peak'
    return  thermal_broadening(E, T) * Breit_Wigner(E, A, Vg, Vg_0, Gamma)

def FanoIntegrand(E, A, Vg, Vg_0, Gamma, qRe, qIm, T):
    'the intergrand for the Coulomb peak'
    return  thermal_broadening(E, T) * Fano(E, A, Vg, Vg_0, Gamma, qRe, qIm)


#Integral:
def G_CP(A, Vg, Vg_0, Gamma, T, offset):
    sampleE =  np.linspace(-10*kB*T-4*h*Gamma,10*kB*T+4*h*Gamma,10000)
    return scipy.integrate.simps(CoulIntegrand(sampleE, A, Vg, Vg_0, Gamma, T), x=sampleE) + offset
    #return np.sign(V) * scipy.integrate.quad(CoulIntegrand,Delta*e,np.inf,args=( Delta*e,gamma, T,V))[0]

def G_Fano(A, Vg, Vg_0, Gamma, qRe, qIm, T, offset):
    sampleE =  np.linspace(-10*kB*T-4*h*Gamma,10*kB*T+4*h*Gamma,10000)
    return scipy.integrate.simps(FanoIntegrand(sampleE, A, Vg, Vg_0, Gamma, qRe, qIm, T), x=sampleE) + offset



def ifitter(Vgate, Vg_0, A, Gamma, T, offset):
    #print(h*Gamma/e/1e-6, 4*kB*T/e/1e-6, Gamma/1e9, T)
    return [G_CP(A=A, Vg=el, Vg_0=Vg_0, Gamma=Gamma, T=T, offset=offset) for el in Vgate]

def ifitter_Fano(Vgate, Vg_0, A, Gamma, qRe, qIm, offset):
    return [G_Fano(A=A, Vg=el, Vg_0=Vg_0, Gamma=Gamma, qRe=qRe, qIm=qIm, offset=offset, T=0.035) for el in Vgate]


mpl.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(1, 1, figsize=[6,5])
testVg = np.linspace(-0.0015,0.0015,500)
for q in [0.0, 2.0, 4.0]:
    testG = [G_Fano(A=0.03, Vg=el, Vg_0=0, Gamma=4.0e9, qRe=-4, qIm=q, offset=0.07, T=0.035) for el in testVg]
    plt.plot(testVg, testG, linewidth=2)

ax.set_xlabel('$V_G$ (V)', fontsize=20)
ax.set_ylabel('G ($2e^2/h$)', fontsize=20)
tick_spacing_x = 0.001
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing_x))
plt.tight_layout()
#fig.savefig("Fano01.pdf", dpi=300, bboxes_inches='tight', pad_inches=0)
```

# Fitting

```python
fname = 'device1_212'
filename = os.path.join(FileDir, fname + '.txt')
df = pd.read_csv(filename, delimiter='\t', header=0)
print(df.head())
Vg_left = 0.00955
Vg_right = 0.01165
df = df[df['ivvi_dac14'] > Vg_left*1e5]
df = df[df['ivvi_dac14'] < Vg_right*1e5]
VPG = df['ivvi_dac14']/1e5
G = df['conductance']
mpl.rcParams.update({'font.size': 14})
fig, ax = plt.subplots(1, 1, figsize=(5,4))
plt.plot(VPG, G, linewidth=2.5)
ax.set_xlim(Vg_left, Vg_right)
#ax.set_ylim(0, 0.35)
ax.set_xlabel('V$_{PG}$ (V)', size=18)
ax.set_ylabel('G (2e$^2$/h)', size=18)
ax.set_title(fname)
tick_spacing_x = 0.0005
tick_spacing_y = 0.1
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing_x))
ax.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing_y))
ax.tick_params(direction='out', length=8, width=1.5, right=False, top=False)

Vgate=VPG.tolist()
Cond=G.tolist()

peaks = coa.detect_peaks(Cond, mph=0.30, mpd=15, threshold=0, edge='both', \
                         kpsh=False, valley=False, show=False, ax=None)
listpeaks = [Vgate[i] for i in peaks]
ax.scatter(listpeaks, [Cond[x] for x in peaks], 16, marker='o', color='r', zorder=100)

Vgates = [Vgate] * len(listpeaks)
cond_peaks = [Cond[i] for i in peaks]
plt.tight_layout()

fit_single_trace = True
if fit_single_trace == True:
    A = 14.7
    Gamma = 1.2e7
    qRe = -1.6
    T = 0.076
    offset = -14.5
    guess = []
    guess += [0.0106] #listpeaks
    guess += [A] * len(listpeaks) #(np.array([A] * len(listpeaks))*max(cond_peaks)/np.array(cond_peaks)).tolist()
    guess += [Gamma] * len(listpeaks)
    guess += [qRe, T, offset]
    
    def nifitter(Vgates, *params):
        r = [0] * len(Vgates[0])
        for i,gate in enumerate(Vgates):
            r = np.array(r)+np.array(ifitter_Fano(gate, params[i], params[i+len(Vgates)], params[i+2*len(Vgates)], params[-3], params[-2], params[-1]))
        r = r - (len(Vgates)-1)*params[-1]
        return r.tolist() #np.array(r).flatten()
    
    popt_list = []
    pcov_list = []
    popt = np.array(guess)
    
    popt, pcov = scipy.optimize.curve_fit(nifitter, Vgates, Cond, p0=guess, xtol = 0.000000001, maxfev = 2500)
    popt_list.append(popt)
    pcov_list.append(pcov)
    
    plt.plot(Vgates[0], nifitter(Vgates, *popt), 'g--')
    plt.show()
```

```python
FileDir = 'data' #relative path to file directory
dirname = os.path.dirname(os.path.abspath('__file__'))
FileDir = os.path.join(dirname, FileDir)
fname = 'device1_4860'
filename = os.path.join(FileDir, fname + '.txt')
df = pd.read_csv(filename, delimiter='\t', \
                 header=0, names=['B', 'PG', 'K1', 'Lx', 'Lphi', 'L_proc', 'G'])
#print(df.head())
#print(df.tail())

df['PG'] = df['PG']/1000*0.01
datapoints_orig = 100 #number of datapoints in gate space
num_fields = 300 #number of field values
Vrange = 0.004
cut_left = 0.0165
cut_right = 0.020
datapoints = round(datapoints_orig*(cut_right-cut_left)/Vrange)
df = df[df['PG'] >= cut_left]
df = df[df['PG'] <= cut_right]

#df['G'] = df['G']/1e6
#Rser=2*(1350+800)+3000 # series resistance for K1
#AC_exc=3.5355e-6
#df['G']=h/e**2/2*df['G']/(AC_exc-df['G']*Rser)
G = []
for i in range(0, num_fields):
    G.append(np.array(df['G'].tolist()[i*datapoints:(i+1)*datapoints]).transpose())
G = G[::-1] #reverse columns if field sweep begins at a larger field
B_start = 0.56
B_end = 0.76
Vg_start = cut_left
Vg_end = cut_right
x_coor = np.linspace(Vg_start, Vg_end, datapoints)
y_coor = np.linspace(B_start, B_end, num_fields)
x, y = np.meshgrid(x_coor, y_coor)
mpl.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(1, 1, figsize=(10,8))
im = ax.pcolor(x, y, np.array(G), cmap=plt.cm.inferno)
cbar = fig.colorbar(im)
cbar.set_label('$G$ (2e$^2$/h)', fontsize = 18)
ax.set_title(fname)
ax.set_xlabel('$V_{PG}$ (V)', fontsize = 20)
ax.set_ylabel('$B$ (T)', fontsize = 20)
ax.set_xlim((Vg_start, Vg_end))
ax.set_ylim((B_start, B_end))
tick_spacing_x = 0.001
tick_spacing_y = 0.05
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing_x))
ax.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing_y))
ax.tick_params(direction='out', length=8, width=1.5, right=False, top=False)

linecut = 0.64 #field in T where line cut is taken
index = y_coor.tolist().index(find_nearest(y_coor, linecut))
plt.axhline(y=y_coor[index], xmin=0, xmax=0.05, linewidth=3.5, color = 'lime')
plt.show()

fig, ax = plt.subplots(1, 1, figsize=(6,4))
ax.plot(x[0], np.array(G)[index].tolist())
ax.set_title(fname)
ax.set_xlabel('$V_{PG}$ (V)', fontsize = 20)
ax.set_ylabel('$G$ (2e$^2$/h)', fontsize = 20)
#ax.set_xlim((Vg_start, Vg_end))
#ax.set_ylim((0, 1.3e-4))
tick_spacing_x = 0.002
#tick_spacing_y = 5e-5
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing_x))
#ax.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing_y))
ax.tick_params(direction='out', length=8, width=2, right=False, top=False)

peaks = coa.detect_peaks(G[index].tolist(), mph=0.16, mpd=20, threshold=0, edge='both', kpsh=False, valley=False, show=False, ax=None)
listpeaks = [x[0][i] for i in peaks]
ax.scatter(listpeaks, [G[index].tolist()[x] for x in peaks], 16, marker='o', color='r')

Vgate=x[0].tolist()
Vgates = [Vgate] * len(listpeaks)
Cond=G[index].tolist()
cond_peaks = [Cond[i] for i in peaks]


fit_single_trace = False
if fit_single_trace == True:
    A = 0.02
    Gamma = 4.0e9
    qRe1 = -4.4
    qIm1 = 1.3
    T = 0.035 #electron temperature is not fitted, change manually in the model above
    offset = 0.07
    guess = []
    guess += listpeaks
    guess += [A] * len(listpeaks) #(np.array([A] * len(listpeaks))*max(cond_peaks)/np.array(cond_peaks)).tolist()
    guess += [Gamma] * len(listpeaks)
    guess += [qRe1, qIm1, offset]
    
    def nifitter(Vgates, *params):
        r = [0] * len(Vgates[0])
        for i,gate in enumerate(Vgates):
            r = np.array(r)+np.array(ifitter_Fano(gate, params[i], params[i+len(Vgates)], params[i+2*len(Vgates)], params[-3], params[-2], params[-1]))
        r = r - (len(Vgates)-1)*params[-1]
        return r.tolist() #np.array(r).flatten()
    
    popt_list = []
    pcov_list = []
    popt = np.array(guess)
    
    popt, pcov = scipy.optimize.curve_fit(nifitter, Vgates, Cond, p0=guess, xtol = 0.000000001, maxfev = 2000)
    popt_list.append(popt)
    pcov_list.append(pcov)
    
    plt.plot(Vgates[0], nifitter(Vgates, *popt), 'g--')
    plt.show()


fit_peaks = True
if fit_peaks == True:
    #Fit all Coulomb peaks using the same electron temperature:
    A = 0.02
    Gamma = 4.0e9
    qRe1 = -4.4
    qIm1 = 1.3
    T = 0.035 #electron temperature is not fitted, change manually in the model above
    offset = 0.07
    guess = []
    guess += listpeaks
    guess += [A] * len(listpeaks) #(np.array([A] * len(listpeaks))*max(cond_peaks)/np.array(cond_peaks)).tolist()
    guess += [Gamma] * len(listpeaks)
    guess += [qRe1, qIm1, offset]
    
    def nifitter(Vgates, *params):
        r = [0] * len(Vgates[0])
        for i,gate in enumerate(Vgates):
            r = np.array(r)+np.array(ifitter_Fano(gate, params[i], params[i+len(Vgates)], params[i+2*len(Vgates)], params[-3], params[-2], params[-1]))
        r = r - (len(Vgates)-1)*params[-1]
        return r.tolist() #np.array(r).flatten()
    
    popt_list = []
    pcov_list = []
    popt = np.array(guess)
    field_values = np.linspace(0.64, 0.685, 67)
    for linecut in field_values:
        print(linecut)
        index = y_coor.tolist().index(find_nearest(y_coor, linecut))
        plt.plot(x_coor, G[index].tolist())
        peaks = coa.detect_peaks(G[index].tolist(), mph=0.16, mpd=20, threshold=0, edge='both', kpsh=False, valley=False, show=False, ax=None)
        listpeaks = [x[0][i] for i in peaks]
        Cond=G[index].tolist()
        cond_peaks = [Cond[i] for i in peaks]
        
        guess = []
        guess += listpeaks
        guess += popt.tolist()[len(listpeaks):]
        
        popt, pcov = scipy.optimize.curve_fit(nifitter, Vgates, Cond, p0=guess, xtol = 0.000000001, maxfev = 2000)
        popt_list.append(popt)
        pcov_list.append(pcov)

    
        #h*Gamma, 4kT, Gamma/GHz, T

        plt.plot(Vgates[0], nifitter(Vgates, *popt), 'g--')
        plt.show()
        plt.figure()
```
